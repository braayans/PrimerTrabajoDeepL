```python
#Librerias usadas
import pandas as pd
from google.colab import files
```

```python
#Lectura del dataset
datos = pd.read_csv("/content/Guadalajara_de_Buga___Impuesto_Predial_2001_-_2005_20240211.csv")
datos.isnull().sum()   #No hay datos NULL
```

```python
#Filtrado por datos con vigencia hasta el año 2023
datos_2023 = datos[datos["VIG. FINAL"] == 2023]
datos_2023
```

#*Transformacion de datos*

```python
#se reemplaza el unico valor de cadena de caracter por un valor entero

datos_2023['ESTRATO SOCIAL'].value_counts()   #SIN INFORMACION no es entero


datos_2023['ESTRATO SOCIAL'] = datos_2023['ESTRATO SOCIAL'].replace({'SIN INFORMACIÓN' : 0 }) #Cambiar SIN INFORMACION POR 0
```

```python
#Se reemplazan los valores de caracteres a binarios


datos_2023['TIPO DE PREDIO'].value_counts()   #Rural o urbano
datos_2023['FUENTE'].value_counts()   #sin informacion o IGAC


datos_2023['TIPO DE PREDIO'] = datos_2023['TIPO DE PREDIO'].replace({'PREDIO RURAL': 0, 'PREDIO URBANO': 1})
datos_2023['FUENTE'] = datos_2023['FUENTE'].replace({'SIN INFORMACIÓN' : 0, 'IGAC': 1 })
```

```python
#Aplicar one hot encoding
datos_2023['DESTINO ECONÓMICO'].value_counts()
datos_2023['CLASE'].value_counts()


datos_2023 = pd.get_dummies(datos_2023, columns = ["DESTINO ECONÓMICO", "CLASE"])
```

#*Eliminar columnas que no se han considerado necesarias*

*   Estado del predio. Todos los valores son iguales
*   Deuda total. Es la suma de dos columnas
*   Ultima novedad. Fecha sin relevancia



```python
datos_2023 = datos_2023.drop(['ESTADO DEL PREDIO', 'DEUDA TOTAL', 'ULTIMA NOVEDAD'], axis = 1)
```

#*Descarga del datase*

```python
datos_2023.to_csv('datos_tarea.csv', index = False)

# Descargar el archivo CSV al sistema local
files.download('datos_tarea.csv')
```
